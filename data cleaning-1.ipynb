```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Data Pre-Processing

# A) for train data

```python
# Reading data
df= pd.read_csv('train_users_2.csv')
```

```python
df.head()
```

```python
df.age.describe()
```

```python
df['age'].fillna(df['age'].mean(),inplace =True)
df.info()
```

```python
df.drop(['date_first_booking'],axis=1,inplace=True)
```

```python
df.first_affiliate_tracked.fillna(value='untracked',inplace=True)
```

```python
df.date_account_created = df.date_account_created.astype('datetime64[ns]')
df['dac_year']=df.date_account_created.dt.year
df['dac_month']=df.date_account_created.dt.month
df['dac_day']=df.date_account_created.dt.day
df.timestamp_first_active = df.timestamp_first_active.astype('datetime64[ns]')
df['tfa_date']=df['timestamp_first_active'].dt.date
df['tfa_time']=df['timestamp_first_active'].dt.time
```

```python
df.drop(['timestamp_first_active','tfa_date','tfa_time','date_account_created','id'],axis=1,inplace=True)
```

```python
nominal_features=['affiliate_channel','affiliate_provider','first_affiliate_tracked','first_browser','first_device_type',
              'gender','language','signup_app','signup_method']
```

```python
from sklearn.preprocessing import LabelEncoder
encoder=LabelEncoder()
encoded=df[nominal_features].apply(encoder.fit_transform)
```

```python
encoded.head()
```

```python
df2=df.drop(nominal_features,axis=1)
```

```python
df=pd.concat([df2,encoded],axis=1)
```

```python
df.to_csv("user training 2 Cleaned.csv", index = False)
```

```python
df.head()
```
